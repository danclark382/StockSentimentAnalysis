```python
with open('twitter_network.json', 'r') as f:
    data = json.load(f)
count = np.array([len(v) for v in data.values()])    
updated_results = copy.deepcopy(data)  
```

```python
# Flat list of all friends
all_friends = [friend for k, v in data.items()
                   for friend in v]
# Get frequency of all friends
c = Counter(all_friends)
# Get 10,000 most common friends
common_friends = list(c.most_common(100))
common_set = {i[0] for i in common_friends}
# Update dictionary to only include values found in the top 100
for k, v in data.items():
    lst = []
    for f in v:
        if f in common_set:
            lst.append(f)
    updated_results[k] = lst
```

```python
accounts, friends = [], []
updated_count = []
idx = 1
for k, v in updated_results.items():
    accounts.append([idx, k])
    updated_count.append(len(v))
    for f in v:
        friends.append([idx, f])
    idx += 1
updated_count = np.array(updated_count)
updated_results = {k:{'friends': v, 'count': len(v)} for k, v in updated_results.items()}
data = {k:{'friends': v, 'count': len(v)} for k, v in data.items()}
```

```python
with open('twitter_accounts.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    for account in accounts:
        writer.writerow([account[0], account[1]])
with open('twitter_friends.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    for friend in friends:
        writer.writerow([friend[0], friend[1]])
```
